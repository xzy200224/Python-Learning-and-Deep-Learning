# 医学图像

```python
import numpy as np
import torch
torch.set_printoptions(edgeitems=2, threshold=50)
```

```python
import imageio
#imageio.volread 则是用于读取体积数据（通常是医学影像等）。它可以读取 DICOM、NIfTI 等格式的数据，并返回一个包含体积数据的 NumPy 数组。
dir_path = "../data/p1ch4/volumetric-dicom/2-LUNG 3.0  B70f-04083"
vol_arr = imageio.volread(dir_path, 'DICOM')
vol_arr.shape
```

```python
# 转tensor浮点型
vol = torch.from_numpy(vol_arr).float()
vol = torch.unsqueeze(vol, 0)   #追加batch

vol.shape
```
